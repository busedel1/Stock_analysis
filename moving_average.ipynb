#### 移動平均とは

$y_t$という時系列データがあるとすると，過去の一定期間$k$との平均を表す時系列データ$z_t$は以下で定義する．

\begin{aligned} z_{t} &=\frac{1}{k+1} \sum_{j=-k}^{0} y_{t+j}, \\ t &=k+1, k+2, \ldots, n \end{aligned}

この$z_t$を移動平均という．
$k$には$20,50,100,200$が多く使用されている．

```python
k=20
y=np.random.rand(100)
v=np.ones(k, dtype="double")/(k)# 畳み込みの重み
z=np.convolve(y,v, mode='same') # 畳み込み
```

```python
plt.plot(y,'r',linewidth=3)
plt.plot(z,'b',linewidth=3)
```
